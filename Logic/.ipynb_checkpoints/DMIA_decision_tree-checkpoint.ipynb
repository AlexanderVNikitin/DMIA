```python
def plot_gain(gain, P, N):
    fig, ax = pylab.subplots()

    plot = pylab.pcolor(gain.T)
    pylab.xlim((0,P))
    pylab.ylim((0,N))

    colorbar = fig.colorbar(plot, ax=ax)
```

## Idea:
Predicates $\phi: X \to \{0, 1\}$

Predicate cover x iff $\phi(x) = 1$

Looking for the most informative predicate.


$P_c$ -- number of c-class objects in train <br>
$p_c = \#\{$objects of c-class in train where $\phi(x) = 1\}$

$N_c$ -- number of not c-class objects in train <br>
$n_c = \#\{$objects of not c-class in train where $\phi(x) = 1\}$

$E_c(\phi, X^l) = \frac{n_c(\phi)}{(p_c(\phi) + n_c(\phi))}$ <br>

$D_c(\phi, X^l) = \frac{p_c(\phi)}{l}$

```python
P = 100
N = 300

def negative_rate(p, n):
    return n * 1.0/(p + n) if p + n != 0 else .0

def positive_rate(p, n):
    return p * 1.0/(p + n) if p + n != 0 else .0
```

```python
def heuristic(P, N, eps, delta):
    gain = np.zeros((P + 1, N + 1))
    for p in range(P + 1):
        for n in range(N + 1):
            gain[p, n] = int(negative_rate(p, n) <= eps and positive_rate(P, p) >= delta)
    return gain
```

```python
gain = heuristic(P, N, eps=0.2, delta=0.5)
plot_gain(gain, P, N)
```

## Statistical way
$h(p, n, P, N) = \frac{C_P^p C_N^n}{C_{N+P}^{n+p}}$<br>
$I_c(\phi) = -ln(h(p_c(\phi), n_c(\phi), P_c(\phi), N_c(\phi)))$

```python
def statistical_gain(p, n, P, N):
    return -np.log(comb(P, p) * comb(N, n) * 1.0/comb(N + P, n + p))

def get_statistical_gain(P, N):
    gain = np.zeros((P + 1, N + 1))
    for p in range(P + 1):
        for n in range(N + 1):
            gain[p, n] = statistical_gain(p, n, P, N)
    return gain

gain = get_statistical_gain(P, N)
plot_gain(gain, P, N)
```

## Entropy

$H(q_0, q_1) = -q_0 * log(q_0) - q_1 * log(q_1)$

Entropy of $X^l$:<br>
$H_{X^l}(P, N) = H(\frac{P}{N + P}, \frac{N}{P + N})$

$H_{X^l}(p, n, P, N) = \frac{p + n}{P + N}H(p, n) + \frac{P + N - p - n}{P + N} H(P - p, N - n)$

$IGain_c = H_{X_l}(P, N) - H_{X_l}(p, n, P, N)$

####Theorem:
$$IGain_c \to \frac{1}{l \ln2} I_c(\phi, X^l), 
l \to \inf$$

```python
def entropy(p1, p2):
    return -p1 * np.log2(p1) - p2 * np.log2(p2)

def H(P, N):
    return entropy(P * 1.0 / (N + P), N * 1.0 / (N + P)) if P + N != 0 else .0

def entropy_gain(p, n, P, N):
    H_train = H(P, N)
    H_splited = ((p + n) * 1.0 / (P + N)) * H(p, n) + ((P + N - p - n) * 1.0 / (P + N)) * H(P - p, N - n)
    return H_train - H_splited

def get_entropy_gain(P, N):
    gain = np.zeros((P + 1, N + 1))
    for p in range(P + 1):
        for n in range(N + 1):
            gain[p, n] = entropy_gain(p, n, P, N)
    return gain

gain = get_entropy_gain(P, N)
plot_gain(gain, P, N)
```

```python
from sklearn.externals.six import StringIO 
def draw_tree(clf):
    dot_data = StringIO()
    tree.export_graphviz(clf, out_file=dot_data) 
    graph = pydot.graph_from_dot_data(dot_data.getvalue())
    return graph.create_png()
```

```python
iris = load_iris()

clf = tree.DecisionTreeClassifier(criterion='entropy')
clf = clf.fit(iris.data, iris.target)
Image(draw_tree(clf), width = 700, height = 700)
```

## Pruning
* <b>pre-pruning</b> -- if informativeness less the $\epsilon$, stop building tree
* <b>post-pruning</b> -- looking on validation dataset. Replace vertex if it shows bad results

```python
from sklearn.datasets.samples_generator import make_blobs
X, y = make_blobs(n_samples=400, centers=4,
                  random_state=0, cluster_std=0.5)
plt.scatter(X[:, 0], X[:, 1], c=y, s=50);
```

```python
def plot_estimator(estimator, X, y):
    estimator.fit(X, y)
    x_min, x_max = X[:, 0].min() - .1, X[:, 0].max() + .1
    y_min, y_max = X[:, 1].min() - .1, X[:, 1].max() + .1
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100),
                         np.linspace(y_min, y_max, 100))
    Z = estimator.predict(np.c_[xx.ravel(), yy.ravel()])

    Z = Z.reshape(xx.shape)
    plt.figure()
    plt.pcolormesh(xx, yy, Z, alpha=0.5)
    
    plt.scatter(X[:, 0], X[:, 1], c=y)
    plt.axis('off')
    plt.axis('off')
    plt.tight_layout()
```

```python
clf = DecisionTreeClassifier(max_depth=125)
plot_estimator(clf, X, y)
```
