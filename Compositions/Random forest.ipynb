## Random forest 
RF is combination of 2 methods: bagging and RSM with decision tree.


```python
from sklearn.datasets.samples_generator import make_blobs
X, y = make_blobs(n_samples=400, centers=4,
                  random_state=0, cluster_std=0.5)
plt.scatter(X[:, 0], X[:, 1], c=y, s=50);
```

```python
def plot_estimator(estimator, X, y):
    estimator.fit(X, y)
    x_min, x_max = X[:, 0].min() - .1, X[:, 0].max() + .1
    y_min, y_max = X[:, 1].min() - .1, X[:, 1].max() + .1
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100),
                         np.linspace(y_min, y_max, 100))
    Z = estimator.predict(np.c_[xx.ravel(), yy.ravel()])

    Z = Z.reshape(xx.shape)
    plt.figure()
    plt.pcolormesh(xx, yy, Z, alpha=0.5)
    
    plt.scatter(X[:, 0], X[:, 1], c=y)
    plt.axis('off')
    plt.axis('off')
    plt.tight_layout()
```

```python
clf_decision = DecisionTreeClassifier(max_depth=125)
clf_rf = RandomForestClassifier(n_estimators=1000)
plot_estimator(clf_rf, X, y)
```

## Why do everybody love RF?

```python
from sklearn.metrics import accuracy_score
from sklearn.cross_validation import train_test_split

#X, y = datasets.make_hastie_10_2(n_samples=25000)
X, y = datasets.make_classification(n_samples = 15000, n_features = 1500)

print X.shape

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 42)


plt.scatter(X[:, 0], X[:, 1], c=y, s=50)
```

```python
xs = []
accuracy_train = []
accuracy_test = []
for n_estimators in xrange(1, 300, 25):
    xs.append(n_estimators)
    if n_estimators % 50 == 1:
        print "Finished %s number of estimators" % str(n_estimators)
    clf_rf = RandomForestClassifier(n_estimators=n_estimators, max_depth=7)
    clf_rf.fit(X_train, y_train)
    
    # For train data
    y_pred = clf_rf.predict(X_train)
    accuracy_train.append(accuracy_score(y_pred, y_train))
    
    # For test data
    y_pred = clf_rf.predict(X_test)
    accuracy_test.append(accuracy_score(y_pred, y_test))
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)

plt.plot(xs, accuracy_train, 'b', xs, accuracy_test, 'r')
plt.show()
```

Usually it doesn't overfit, so it's easy to find suitable params.

Suppose we have M trees in Random forest:<br>
$Var(RF) = \frac{1}{M}\sum\limits_{i = 1}^{M}(Var(T_i)) = \frac{1}{M^2}\sum\limits_{i = 1}^{M}\sum\limits_{i = 1}^{M}(Cov(T_i, T_j)) =  \frac{1}{M^2}\sum\limits_{i = 1}^{M}(\sum\limits_{i = 1, i \neq j}^{M}Cov(T_i, T_j) + Var(T_i)) = \frac{1}{M^2}\sum\limits_{i = 1}^{M}(M  \rho  \sigma^2 + \sigma^2)$<br>
$\rho \to 0$, random features subsampling<br>
